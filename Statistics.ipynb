# Introduction  
**What is statistics?**

Statistics is the science of **collecting, analyzing, interpreting, and presenting data**. It is used to make sense of the world around us and to draw conclusions from data. Statistics is used in many different fields, including business, science, government, and medicine.

**The four main steps of statistics are:**

1. **Data collection:** This involves gathering data from a population or sample.  
2. **Data analysis:** This involves organizing, summarizing, and describing the data.  
3. **Data interpretation:** This involves drawing conclusions from the data.  
4. **Data presentation:** This involves communicating the results of the analysis to others.

**Types of statistics:**

There are two main types of statistics:

**Descriptive Statistics:** These methods summarize and describe the main features of a dataset. Examples include mean, median, mode, range, and standard deviation. Descriptive statistics give you a snapshot of what your data looks like.
     - *Example:* If you have a list of test scores (80, 85, 90, 95), the average (mean) could be a descriptive statistic to tell you the typical score.

**Inferential Statistics:** These methods make inferences or predictions about a population based on a sample of data. It involves drawing conclusions from data and assessing the likelihood of those conclusions being true.
     - *Example:* If you survey a group of people and find that 70% prefer product A, you might use inferential statistics to estimate that the whole population would also have a similar preference.

In summary, descriptive statistics help you understand your data, inferential statistics help you make predictions or inferences about a larger group.

**Why is statistics important?**

Statistics is important because it helps us to make informed decisions. By understanding data, we can better understand the world around us and make decisions that are based on evidence rather than opinion.

**Here are some of the benefits of using statistics:**

* **Helps us to identify trends and patterns:** Statistics can help us to identify trends and patterns in data that we might not otherwise see.  
* **Allows us to make predictions:** Statistics can be used to make predictions about future events.

# Population and Sample

## Population:

**Definition:** The population is the entire group that is the subject of the study or the group in which we are interested in. It includes all individuals, objects, or events that share a common characteristic.

**Example:** If we want to study the average height of all adults in a country, the population would be all the adults in that country.

**Symbol:** N

## Sample:

**Definition:** A sample is a subset of the population. It's a smaller group chosen to represent the entire population.

**Example:** Instead of measuring the height of every adult in the country, we might select a sample of 500 adults and use their heights to make inferences about the average height of the entire population.

**Symbol:** n

### Why Use a Sample?

- **Practicality:** It's often impractical or impossible to study the entire population due to time, cost, or logistical constraints.

- **Accuracy:** With proper sampling methods, a well-chosen sample can provide accurate insights into the population.

### Types of Sampling Techniques

1. **Simple Random Sampling:**
   - **Definition:** Every individual in the population has an equal chance of being selected.
   - **Example:** Assigning a unique number to each adult and using a random number generator to select the sample.
   

2. **Stratified Sampling:**
   - **Definition:** The population is divided into subgroups (strata), and individuals are randomly selected from each subgroup. These subgroups are non-overlapping group.
   - **Example:** If studying heights, the strata might be different age groups, and a sample is randomly taken from each group.
       OR I divide my population into two subgroup Male and Female, so i can choose any subgroup and take my sample from that group.
       

3. **Systematic Sampling:**
   - **Definition:** Selecting every kth individual from the population after a random starting point.
   - **Example:** Selecting every 10th person in a list after randomly choosing a starting point.
   

4. **Cluster Sampling:**
   - **Definition:** The population is divided into clusters, and entire clusters are randomly selected.
   - **Example:** Dividing a city into neighborhoods, randomly selecting neighborhoods, and measuring everyone in the chosen neighborhoods.
   

5. **Convenience Sampling**
    - **Definition:** Voluntary response sampling, sampling from people who can voluntary response
    - **Example:** Surveying about datascience, so only data scientist can voluntary fill the survey or i only can ask datascientist to voluntare of filling the survey
   

### When to Use Different Sampling Techniques?

- **Simple Random Sampling:** Useful when the population is homogeneous, and each individual is similar. Example: Studying the average age of students in a school.

- **Stratified Sampling:** Beneficial when the population has distinct subgroups. Example: Analyzing the preferences of different age groups in a community or male or female.

- **Systematic Sampling:** Convenient when there's an ordered list of the population. Example: Sampling every 5th product off a production line.

- **Cluster Sampling:** Suitable when the population naturally divides into clusters. Example: Studying the health of students in randomly selected schools.

Understanding populations, samples, and different sampling techniques is crucial for obtaining reliable and meaningful results in statistical analyses.

# Measure of Central Tendency

## Mean:

**Definition:** The mean is the average of a set of values in a population. It is calculated by summing up all the values and dividing by the total number of size.

**Symbol:** Symbol MU for mean of population and X(bar) for mean of sample: 

**Formula:** for population: Mu = Î£X/N.

**Formula:** for sample:     Xbar = Î£X/n

X is data and Î£X is summation of X, means Sum of all X (values)

N and n is Total size or total count of X

**Example:** Consider a set of exam scores in a population: 85, 90, 92, 88, and 95. The mean is \({85 + 90 + 92 + 88 + 95}/{5} = 90 \).

**Use Cases:**
- When the data is numerical and follows a normal distribution.
- Suitable for interval or ratio data.
- In case of outliers  Mean value is affected, not accurate

## Median:

**Definition:** The median is the middle value when the data is arranged in ascending or descending order. If there is an odd number of values, the median is the middle value. If there is an even number, it's the average of the two middle values.

**Formula for Odd:** just take the middle value

**Formula for Even:** sum the two middle value then divide by 2

**Example:** For the set of exam scores in a population: 85, 90, 92, 88, 95, 

sort: 85,88,90,92,95

the median is 90.

**Use Cases:**
- When dealing with skewed data or outliers.
- Ordinal or interval data.

## Mode:

**Definition:** The mode is the value that appears most frequently in a set of data.

**Example:** In the set of exam scores in a population: 85, 90, 92, 88, 95, the mode is not applicable as there is no repeated value.

**Use Cases:**
- Categorical data or nominal data.
- Can be used alongside mean and median for a complete picture.

## Choosing Between Mean, Median, and Mode:

- **Symmetric Data:** If the data is symmetrically distributed, the mean, median, and mode are approximately equal.
  
- **Skewed Data:** In skewed data, the median is often a better measure of central tendency as it is less influenced by extreme values.

- **Outliers:** If there are outliers in the data, the median might be preferred over the mean.

- **Categorical Data:** For categorical data, the mode is more appropriate.

- **Normal Distribution:** In a normal distribution, all three measures are often suitable, but the mean is frequently used due to its mathematical properties.

In summary, the choice between mean, median, and mode depends on the nature of the data and the specific characteristics of the distribution. Understanding the data distribution and the potential impact of outliers can guide the selection of the most appropriate measure of central tendency.

# Variance and Standard Deviation

Variance and standard deviation are two ways to measure how much a set of data values are spread out from their average value.


## Variance

variance is a measure of how spread out or dispersed a set of data points are from the average (mean) value. It gives you an idea of how much individual data points differ from the overall average.

If the variance is low, it means that most data points are close to the average. On the other hand, a high variance indicates that the data points are more spread out from the average. In essence, variance quantifies the extent to which each data point in a dataset deviates from the mean. 

**Formula**

Var = $Î£(x-xbar)^2 / (n-1)$



formula for Sample variance is:

Var = Î£(x-xbar)^2 / (n-1)
where:

- \\(Var\\) is the variance
- \\(x\\) is a sample data value
- \\(\\xbar\\) is the mean of the sample data
- \\(n\\) is the number of data values
- \\(n-1\\) also called bassels correction and degree of freedom
## Standard Deviation

Standard deviation is the square root of the variance.

it is a measure that indicates how much a value should deviate from the mean (average) of the data.

In a graphical representation, values that are farthest from the mean are positioned:
- To the right side of the graph as `mean + std + std...`
- To the left side of the graph as `mean - std - std...`



The formula for standard deviation is:

$\sqrt(Var)$

let's add an example to illustrate the concepts of variance and standard deviation:

### Example:

Suppose we have the following daily temperatures (in degrees Celsius) recorded over a week: \[20, 22, 18, 25, 23, 21, 19\]

#### Step 1: Calculate the Mean

xbar = {20 + 22 + 18 + 25 + 23 + 21 + 19}/{7} = {148}/{7} = 21.14 ]

#### Step 2: Calculate the Squared Differences and Variance
**Formula :** 

Sample Variance:

Var = Î£(x-xbar)^2 / (n-1)

***note : For population variance will be only 'N' and for sample var = 'n-1'***

x = \[20, 22, 18, 25, 23, 21, 19\]

xbar = 21.14

n = 7

Var = 5.8096

#### Step 3: Calculate the Standard Deviation

$\sqrt{5.8096} \approx 2.41$

### Interpretation:

- The mean temperature is approximately \(21.14\) degrees Celsius.
- The variance is \(5.8096\), indicating the average squared difference of each temperature from the mean.
- The standard deviation is approximately \(2.41\), providing a more interpretable measure of the spread.

### Application:

- A higher standard deviation suggests that the daily temperatures fluctuate around the Mean(avg) by about \(2.41\) degrees on average.
- This information can be valuable for understanding the variability in daily temperatures, especially if the data were collected for planning purposes or to assess climate patterns.

In this example, variance and standard deviation help quantify the extent to which individual daily temperatures deviate from the weekly average, providing insights into the consistency or variability in the recorded temperatures.


````python
import numpy as np

# Given daily temperatures
temperatures = np.array([20, 22, 18, 25, 23, 21, 19])

#lets manually find the variance from what we learned (the hard way).
#Formula: var = (Sum of (data - mean of data)^2)/n-1  --> n is size

n = len(temperatures) #n
tmean = np.mean(temperatures) #mean of data
deviation = [(x-tmean)**2 for x in temperatures]

var = sum(deviation)/(n-1)
print('variance: ',var)


#Easy way just:

print('population Variance: ',np.var(temperatures)) 
#numpy np.var give us population variance, we need to define bassel's correction for sample variance

print('Sample Variance: ',np.var(temperatures, ddof=1)) #ddof=1 --> n-1

# standard deviation
print('Standard deviation: ',np.sqrt(var))
````

````python
#on the graph

import numpy as np
import matplotlib.pyplot as plt

# Given daily temperatures
temperatures = np.array([20, 22, 18, 25, 23, 21, 19])

# Calculate mean, variance, and standard deviation
mean_temperature = np.mean(temperatures)
variance_temperature = np.var(temperatures, ddof=1)
std_dev_temperature = np.sqrt(variance_temperature)

# Create a plot
plt.figure(figsize=(8, 5))

# Plot daily temperatures
plt.scatter(range(1, len(temperatures) + 1), temperatures, label='Daily Temperatures')

# Plot mean temperature
plt.axhline(y=mean_temperature, color='g', linestyle='--', label='Mean Temperature')


# Plot one standard deviation above and below the mean
plt.axhline(y=mean_temperature + std_dev_temperature, color='b', linestyle='--', label='+1 Std Dev')
plt.axhline(y=mean_temperature - std_dev_temperature, color='r', linestyle='--', label='-1 Std Dev')

#axline draw horizontal line

# Set plot labels and title
plt.xlabel('Day')
plt.ylabel('Temperature (degrees Celsius)')
plt.title('Daily Temperatures with Mean and Standard Deviation')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

````

## Quartile

The 1st quartile \(Q1\) is the 25th percentile, \(Q3\) is the 75th percentile, and \(Q2\) is the median (50th percentile).

**Interquartile Range (IQR)**

\[ IQR = Q3 - Q1 \]

We can use quartiles to find outliers and also used in a box plot to find outliers in the data. The range for identifying outliers is determined by the lower fence (\(Q1 - 1.5 \times IQR\)) and the upper fence (\(Q3 + 1.5 \times IQR\)).

````python
iqr= q3-q1  #Inter Quartile Range
iqr
````

````python
Lrange = q1-(1.5*iqr) #lower fence
Lrange
````

## Probability Mass Function

PMF stands for Probability Mass Function. It is a mathematical function that 
describes the probability distribution of a discrete random variable.
The PMF of a discrete random variable assigns a probability to each possible value 
of the random variable. The probabilities assigned by the PMF must satisfy two 

conditions:
- The probability assigned to each value must be non-negative (i.e., greater than or equal to zero).
- The sum of the probabilities assigned to all possible values must equal 1.

Example : probability of dice possible values

The cumulative distribution function (CDF) F(x) describes the probability that a 
random variable X with a given probability distribution will be found at a value less 
than or equal to x

 F(x) = P(X <= x)

## Probability Density Function
PDF stands for Probability Density Function. It is a mathematical function that 
describes the probability distribution of a continuous random variable.

A probability density function (PDF) describes the likelihood of a continuous random variable falling within a certain range. Unlike discrete probabilities, where each outcome has a specific probability, PDF deals with continuous variables, so the probability at any specific point is often zero. Instead, the area under the PDF curve over a given range represents the probability of the variable falling within that range. In essence, it provides a way to understand the likelihood of different outcomes in a continuous distribution.



1. **Why Probability Density and not Probability?**
   - Probability density is used for continuous random variables because they can take on an infinite number of values within a range. Unlike discrete variables, where each value has a specific probability, for continuous variables, the probability at any single point is usually zero. Probability density helps us understand the likelihood of a variable falling within a certain range rather than at a specific point.

2. **What does the area of this graph represent?**
   - The area under the graph of a probability density function represents the probability of the random variable falling within a certain range. In simple terms, if you want to know the probability of a continuous random variable being between two values, you find the area under the curve between those two values. The larger the area, the higher the probability.

3. **How to calculate Probability then?**
   - To calculate the probability of a continuous random variable falling within a specific range, you need to find the area under the curve of its probability density function within that range. This is typically done using calculus techniques, such as integration. So, you integrate the PDF over the range of interest to find the probability. For example, if you want to find the probability of a variable \( X \) falling between \( a \) and \( b \), you would integrate the PDF from \( a \) to \( b \):
   
    P(a â¤ X â¤ b) = â«[a, b] f(x) dx
    
- `P(a â¤ X â¤ b)` represents the probability of the variable \( X \) falling between \( a \) and \( b \).
- `â«[a, b]` denotes the integral from \( a \) to \( b \), indicating the process of finding the area under the curve.
- `f(x)` represents the probability density function of the variable \( X \), which defines the distribution of probabilities across the range of values of \( X \).
- `dx` represents the differential element of \( x \), indicating the variable with respect to which the integration is performed.


### Technique to find probability density
#### Density Estimation
 Density estimation is a statistical technique used to estimate the probability 
density function (PDF) of a random variable based on a set of observations or data. 
In simpler terms, it involves estimating the underlying distribution of a set of data 
points.

 Density estimation can be used for a variety of purposes, such as hypothesis testing, data analysis, and data visualization. It is particularly useful in areas such 
as machine learning, where it is often used to estimate the probability distribution of input data or to model the likelihood of certain events or outcomes.

 There are various methods for density estimation, including parametric and non parametric approaches. Parametric methods assume that the data follows a specific probability distribution (such as a normal distribution), while non parametric methods do not make any assumptions about the distribution and instead estimate it directly from the data.

 Commonly used techniques for density estimation include kernel density estimation (KDE), histogram estimation, and Gaussian mixture models (GMMs). 
The choice of method depends on the specific characteristics of the data and the intended use of the density estimate.

#### Parametric Density Estimation
 Parametric density estimation is a method of estimating the probability density 
function (PDF) of a random variable by assuming that the underlying distribution 
belongs to a specific parametric family of probability distributions, such as the

# Normal Distribution

### Definition

Normal distribution, also known as Gaussian distribution, is a probability distribution that 
is commonly used in statistical analysis. It is a continuous probability distribution that is 
symmetrical around the mean, with a bell-shaped curve.



- Tail is infinite 
- Lots of points near the mean and very few far away

- The mean, median, and mode are all equal and located at the center.
- About 68% of the data falls within one standard deviation of the mean.
- About 95% falls within two standard deviations.
- About 99.7% falls within three standard deviations.
- 68%,95%,99.7% is known as empirical rule. It is a statistical guideline that describes the percentage of data that falls within certain standard deviations from the mean in a normal distribution


### Why is it so important?

 Commonality in Nature: Many natural phenomena follow a normal distribution, such as 
the heights of people, the weights of objects, the IQ scores of a population, and many 
more. Thus, the normal distribution provides a convenient way to model and analyse such 
data.

i.e. Most of the people height are around the average(mean) and very less are shorter and taller then average

**Denoted by**

$\mathcal{N}(\mu, \sigma^2)$ or $\text{Norm}(\mu, \sigma^2)$


### PDF Equation of Normal Distribution

$ f(x | \mu, \sigma) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x - \mu)^2}{2\sigma^2}} $

###  Standard Normal Variate

A Standard Normal Variate(Z) is a standardized form of the normal distribution with mean = 0 and standard deviation = 1.

Standardizing a normal distribution allows us to compare different distributions with each 
other, and to calculate probabilities using standardized tables or software such as Z-Table.

**How to transform a normal distribution to Standard Normal Variate**

The formula to transform a normal distribution to a standard normal variate (z-score) is:

$z = \frac{x - \mu}{\sigma}$

- **Procedure**: 
  - Take each data point \( x \), subtract the mean $\mu $, and divide by the standard deviation $\sigma $.
  - The resulting value \( z \) is called the z-score.
  - The z-score indicates how many standard deviations an observation is from the mean.

- **Interpretation**:
  - A positive z-score indicates that the data point is above the mean, while a negative z-score indicates it is below the mean.
  - A z-score of 0 means the data point is exactly at the mean.
  - Larger absolute values of z indicate data points farther from the mean.

- **Using z-table**:
  - A z-table provides the cumulative probability of a standard normal distribution up to a given z-score.
  - By looking up the z-score in the z-table, we can find the cumulative probability, which represents the proportion of observations below that z-score.
  
  **Benefits of Transforming to Standard Normal Variate**

Transforming data to standard normal variates (z-scores) has several benefits:

1. **Standardization**: It allows for the comparison of observations from different normal distributions on a common scale.
   
2. **Interpretation**: Z-scores provide a clear interpretation of how many standard deviations a data point is from the mean, making it easier to understand the relative position of individual observations within the distribution.

3. **Statistical Analysis**: Z-scores are often used in statistical analysis, such as hypothesis testing and confidence interval estimation, as they simplify calculations and facilitate standardization of results across different datasets.

4. **Probabilistic Inference**: Z-scores can be used to compute probabilities associated with specific values or ranges of values in a normal distribution using z-tables or statistical software, enabling probabilistic inference about the data.

# Skewness and Kurtosis

### what is skewness?

A normal distribution is a bell-shaped, symmetrical distribution with a specific 
mathematical formula that describes how the data is spread out. Skewness indicates that 
the data is not symmetrical, which means it is not normally distributed.

Skewness is a measure of the asymmetry of a probability distribution. It is a statistical 
measure that describes the degree to which a dataset deviates from the normal 
distribution.

In a symmetrical distribution (normal), the mean, median, and mode are all equal. In contrast, in a 
skewed distribution, the mean, median, and mode are not equal, and the distribution 
tends to have a longer tail on one side than the other.

Skewness can be positive, negative, or zero. A positive skewness means that the tail of 
the distribution is longer on the right side, while a negative skewness means that the tail 
is longer on the left side. A zero skewness indicates a perfectly symmetrical distribution.



The greater the skew the greater the distance between mode, median and mean
- In positive skewness mode < median < mean. Mean is greater then median
- In negative skewness mean < median < mode. Median is grater then mean

The formula for skewness is:

$$g=\frac{n\sum_{i=1}^n\left(\frac{x_i-\bar{x}}{s}\right)^3}{(n-1)(n-2)}$$

where $x_i$ is the $i$th observation, $\bar{x}$ is the mean, $n$ is the number of observations, and $s$ is the standard deviation.

As we have power 3 in the formula it has 3rd moment in statistics

A moment is a statistical measure that describes the shape and characteristics of a probability distribution. It is calculated as the expected value of a power of the deviation of a random variable from a fixed point, usually the mean. The higher the power, the higher the order of the moment.

The most common moments are:

The first moment (power of 1), which is the mean. It measures the central location of the distribution.

The second moment(Power of 2), which is the variance. It measures the spread or dispersion of the distribution.

The third moment(power of 3), which is the skewness. It measures the asymmetry or lack of symmetry of the distribution.

The fourth moment(power of 4), which is the kurtosis. It is a measure of the tailedness of a distribution, how fat and thin are tails.


### What is Kurtosis?

Kurtosis is the 4th statistical moment. In probability theory and statistics, kurtosis
(meaning "curved, arching") is a measure of the "tailedness" of the probability
distribution of a real-valued random variable. Like skewness, kurtosis describes a
particular aspect of a probability distribution.
Tailedness is how often outliers occur

For Example: If a distribution has high kurtosis (fat tail), it means that there is a higher likelihood of
extreme events occurring, either positive or negative, compared to a normal
distribution.

The formula for sample kurtosis is:

$$g=\frac{n(n+1)\sum_{i=1}^n\left(\frac{x_i-\bar{x}}{s}\right)^4-3(n-1)\left(\sum_{i=1}^n\left(\frac{x_i-\bar{x}}{s}\right)^2\right)^2}{(n-1)(n-2)(n-3)}$$

where $x_i$ is the $i$th observation, $\bar{x}$ is the sample mean, $n$ is the sample size, and $s$ is the sample standard deviation.

**Practical Use-case**

In finance, kurtosis risk refers to the risk associated with the possibility of extreme
outcomes or "fat tails" in the distribution of returns of a particular asset or portfolio.
If a distribution has high kurtosis, it means that there is a higher likelihood of
extreme events occurring, either positive or negative, compared to a normal
distribution.
In finance, kurtosis risk is important to consider because it indicates that there is a
greater probability of large losses or gains occurring, which can have significant
implications for investors. As a result, investors may want to adjust their investment
strategies to account for kurtosis risk.

**Excess Kurtosis & Types**

Excess kurtosis is the tailedness of a distribution relative to a normal distribution.
It is calculated by subtracting 3 from the sample kurtosis coefficient (from above formula sample kurtosis).

Distributions with medium kurtosis (medium tails) are mesokurtic is also a normal distribution. Distributions with low kurtosis (thin tails) are platykurtic. Distributions with high kurtosis (fat tails) are leptokurtic






# QQ Plot

**How to find if a given distribution is normal or not?**
1. Visual inspection: One of the easiest ways to check for normality is to visually inspect a histogram or a density plot of the data. A normal distribution has a bell-shaped curve, which means that the majority of the data falls in the middle, and the tails taper off symmetrically. If the distribution looks approximately bell-shaped, it is likely to be normal.


2. QQ Plot: Another way to check for normality is to create a normal probability plot (also known as a Q-Q plot) of the data. A normal probability plot plots the observed data against the expected values of a normal distribution. If the data points fall along a straight line, the distribution is likely to be normal.


3. Statistical tests: There are several statistical tests that can be used to test for normality, such as the Shapiro-Wilk test, the Anderson-Darling test, and the Kolmogorov-Smirnov test. These tests compare the observed data to the expected values of a normal distribution and provide a p-value that indicates whether the data is likely to be normal or not. A p-value less than the significance level (usually 0.05) suggests that the data is not normal.

**What is a QQ Plot and how is it plotted?**
A QQ plot (quantile-quantile plot) is a graphical tool used to assess the similarity of
the distribution of two sets of data. It is particularly useful for determining whether a
set of data follows a normal distribution.

In a QQ plot, the quantiles of the two sets of data are plotted against each other. The
quantiles of one set of data are plotted on the x-axis, while the quantiles of the other
set of data are plotted on the y-axis. If the two sets of data have the same
distribution, the points on the QQ plot will fall on a straight line. If the two sets of
data do not have the same distribution, the points will deviate from the straight line.

## Power Law Distribution

### Definition:

In statistics, a power law distribution describes a phenomenon where the probability of an event decreases as a **power of its size**. In simpler terms, **a few things are much more likely to happen than many others**. This leads to a skewed distribution with a "long tail" of rare but impactful events.

### Daigram:


distribution exhibit a long tail, indicating that a few large values dominate the distribution.

### Imagine the Apples:

Think of a basket overflowing with apples. Instead of a bell-shaped curve with most apples average-sized, a power law scenario would have mostly **tiny apples**, with just a **few colossal giants** dominating the scene. This illustrates the core principle: **a small number having a disproportionately large influence**.

### Beyond Apples: Real-World Examples:

* **City sizes:** A few megacities have vastly more people than most smaller towns.
* **Earthquakes:** Most are minor tremors, but a few are devastating giants.
* **Word usage:** A few common words make up most of our speech, while many are rarely used.
* **Website popularity:** A handful of websites attract tons of traffic, while countless others struggle.
* **Income distribution:** A small percentage of individuals holds a significant portion of wealth.

### Key Characteristics:

1. **Few Big, Many Small:** The distribution is **uneven**, with a few extreme outliers on one end (the giants) and many smaller occurrences on the other (the tiny apples).
2. **Fat Tail:** Unlike the bell curve, the power law has a **stretched-out tail** towards the extreme values, emphasizing the dominance of the "big players."
3. **Scale Invariance:** Often, power laws exhibit **scale invariance**, meaning the same pattern holds true even if you magnify or shrink the entire distribution.

### Remember:

Power laws don't imply **everything** is extreme. It's about recognizing that **a select few** often hold immense influence within a system, while the majority contributes in smaller ways.

So, the next time you encounter a scenario with seemingly outsized players, consider the power law. It might hold the key to understanding the hidden giants shaping our world!

````python
#CODE
import numpy as np
import matplotlib.pyplot as plt

xm = 2
alpha = 1.16

x = stats.pareto.rvs(b=alpha, scale=xm, size=1000)  #data

x_value = np.linspace(x.min(),x.max(),100)  #observed value

y_value = (alpha*xm**alpha)/(x_value*(alpha+1))

plt.plot(x_value,y_value)
````

## Central Limit Theorem

The Central Limit Theorem (CLT) states that the distribution of the sample means of a large
number of independent and identically distributed random variables will approach a normal
distribution, regardless of the underlying distribution of the variables.
The conditions required for the CLT to hold are:  

**SIMPLY**  
The Central Limit Theorem (CLT) is a statistical principle stating that the distribution of Sample Means from a sufficiently large sample, regardless of the shape of the original population distribution, will be approximately normal distribution or guassion distribution.


## Key Points:
1. **Large Sample Requirement:**
   - CLT applies when working with a sufficiently large sample size. As a rule of thumb, a sample size of 30 or more is often considered large enough.

2. **Sample Mean Distribution:**
   - If you take multiple random samples from any population, calculate the mean of each sample, and plot the distribution of these sample means, the resulting distribution will be approximately normal distribution, even if the original population distribution is not.

3. **Independence Assumption:**
   - The samples should be drawn independently. Each observation's selection should not influence the selection of another observation.

   - The CLT is powerful because it allows us to make inferences about population parameters, especially the mean, without being concerned about the shape of the original population distribution.
   
5. The sample is drawn from a finite population or an infinite population with a finite variance (should be finite).

# Hypothesis Testing in Inferential Statistics

Hypothesis testing is a method in inferential statistics that involves drawing conclusions or making assumptions about a population based on a sample. It aims to determine whether a given hypothesis or assumption about the population is true or false.

**1. Null hypothesis (H0):**  
In simple terms, the null hypothesis is a statement that assumes there is no significant
effect or relationship between the variables being studied. It serves as the starting point
for hypothesis testing and represents the status quo or the assumption of no effect until
proven otherwise. The purpose of hypothesis testing is to gather evidence (data) to either
reject or fail to reject the null hypothesis in favour of the alternative hypothesis, which
claims there is a significant effect or relationship.

**2. Alternative hypothesis (H1 or Ha):**  
The alternative hypothesis, is a statement that contradicts the null hypothesis and claims
there is a significant effect or relationship between the variables being studied. It
represents the research hypothesis or the claim that the researcher wants to support
through statistical analysis.

**Important Points**

How to decide what will be Null hypothesis and what will be Alternate
Hypothesis(Typically the Null hypothesis says nothing new is happening)

We try to gather evidence to reject the null hypothesis

It's important to note that failing to reject the null hypothesis doesn't necessarily mean
that the null hypothesis is true; it just means that there isn't enough evidence to support
the alternative hypothesis.

5. State the relevant test statistic
6. Conduct the test

## Type I error and Type II error

In hypothesis testing, there are two types of errors that can
occur when making a decision about the null hypothesis: 

**Type-I** (False Positive) error occurs when the sample results,
lead to the rejection of the null hypothesis when it is in fact
true.
In other words, it's the mistake of finding a significant effect or
relationship when there is none. The probability of committing
a Type I error is denoted by Î± (alpha), which is also known as
the significance level. By choosing a significance level,
researchers can control the risk of making a Type I error.

**Type-II** (False Negative) error occurs when based on the
sample results, the null hypothesis is not rejected when it is in
fact false.
This means that the researcher fails to detect a significant
effect or relationship when one actually exists. The probability
of committing a Type II error is denoted by Î² (beta).
Trade-off between Type 1 and Type 2 errors

**One-sided (one-tailed) test**: A one-sided test is used when the researcher is interested in
testing the effect in a specific direction (either greater than or less than the value specified in
the null hypothesis). The alternative hypothesis in a one-sided test contains an inequality
(either ">" or "<").
Example: A researcher wants to test whether a new medication increases the average
recovery rate compared to the existing medication.

**Two-sided (two-tailed) test**: A two-sided test is used when the researcher is interested in
testing the effect in both directions (i.e., whether the value specified in the null hypothesis is
different, either greater or lesser). The alternative hypothesis in a two-sided test contains a
"not equal to" sign (â ).
Example: A researcher wants to test whether a new medication has a different average
recovery rate compared to the existing medication.
The main difference between them lies in the directionality of the alternative hypothesis and
how the significance level is distributed in the critical regions.

## Advantages and Disadvantages of one and two tailed test

**Two-tailed test (two-sided):**  
**Advantages:**  
1. Detects effects in both directions: Two-tailed tests can detect effects in both directions,
    which makes them suitable for situations where the direction of the effect is uncertain or
    when researchers want to test for any difference between the groups or variables.

2. More conservative: Two-tailed tests are more conservative because the significance level
    (Î±) is split between both tails of the distribution. This reduces the risk of Type I errors in
    cases where the direction of the effect is uncertain.

**Disadvantages:**  
1. Less powerful: Two-tailed tests are generally less powerful than one-tailed tests because
    the significance level (Î±) is divided between both tails of the distribution. This means the
    test requires a larger effect size to reject the null hypothesis, which could lead to a higher
    risk of Type II errors (failing to reject the null hypothesis when it is false).

2. Not appropriate for directional hypotheses: Two-tailed tests are not ideal for cases where
    the research question or hypothesis is directional, as they test for differences in both
    directions, which may not be of interest or relevance.

**One-tailed test (one-sided):**  
**Advantages:**  
1. More powerful: One-tailed tests are generally more powerful than two-tailed tests, as the
    entire significance level (Î±) is allocated to one tail of the distribution. This means that the
    test is more likely to detect an effect in the specified direction, assuming the effect exists.

2. Directional hypothesis: One-tailed tests are appropriate when there is a strong theoretical
    or practical reason to test for an effect in a specific direction.
    
**Disadvantages:**  
1. Missed effects: One-tailed tests can miss effects in the opposite direction of the specified
    alternative hypothesis. If an effect exists in the opposite direction, the test will not be
    able to detect it, which could lead to incorrect conclusions.

2. Increased risk of Type I error: One-tailed tests can be more prone to Type I errors if the
    effect is actually in the opposite direction than the one specified in the alternative
    hypothesis.

# P-Value

A p-value is a number that tells you how likely it is that your data could have happened by chance, assuming that there is no real effect or difference between the groups you are comparing. The smaller the p-value, the more evidence you have to reject the null hypothesis, which is the assumption that there is no effect or difference.  
In simple words p-value is a measure of the strength of the evidence against the Null
Hypothesis that is provided by our sample data.  
P-Value is the area of rejection

## Interpreting P-Value

Usually we consider significance value as 0.05 with 95 confidence inteval

**Without significance value**  
**Very small p-values** (e.g., p < 0.01) indicate strong evidence against the null hypothesis,
suggesting that the observed effect or difference is unlikely to have occurred by chance
alone. (reject null hypothesis)

**Small p-values** (e.g., 0.01 â¤ p < 0.05) indicate moderate evidence against the null
hypothesis, suggesting that the observed effect or difference is less likely to have

**Very large p-values** (e.g., p â¥ 0.1) indicate weak or no evidence against the null
hypothesis, suggesting that the observed effect or difference is likely to have occurred by

# T-tests

A t-test is a statistical test used in hypothesis testing to compare the means of two samples or
to compare a sample mean to a known population mean. The t-test is based on the t-
distribution, which is used when the population standard deviation is unknown and the
sample size is small.

**There are three main types of t-tests:**  
**One-sample t-test:** The one-sample t-test is used to compare the mean of a single sample to a
known population mean. The null hypothesis states that there is no significant difference
between the sample mean and the population mean, while the alternative hypothesis states
that there is a significant difference.

**Independent two-sample t-test:** The independent two-sample t-test is used to compare the
means of two independent samples. The null hypothesis states that there is no significant
difference between the means of the two samples, while the alternative hypothesis states that
there is a significant difference.

**Paired t-test (dependent two-sample t-test):** The paired t-test is used to compare the means of
two samples that are dependent or paired, such as pre-test and post-test scores for the same
group of subjects or measurements taken on the same subjects under two different
conditions. The null hypothesis states that there is no significant difference between the
means of the paired differences, while the alternative hypothesis states that there is a
significant difference.

# Chi-Square test

The chi-square test, denoted by ÏÂ², is a fundamental statistical tool employed to analyze categorical data. It helps researchers uncover potential relationships or significant differences between categorical variables by comparing observed frequencies (actual data) with expected frequencies (based on a null hypothesis).

**Two types of Chi-Sqaure test**

***The chi-square goodness of fit test*** is used to test whether the frequency distribution of a categorical variable is different from your expectations. (difference)  
***The chi-square test of independence*** is used to test whether two categorical variables are related to each other. (relationship)

Pearsonâs chi-square (Î§2) tests, often referred to simply as chi-square tests, are among the most common ***nonparametric tests***. Nonparametric tests are used for data that donât follow the assumptions of parametric tests, especially the assumption of a normal distribution.

If you want to test a hypothesis about the distribution of a categorical variable youâll need to use a chi-square test or another nonparametric test. Categorical variables can be nominal or ordinal and represent groupings such as species or nationalities. Because they can only have a few specific values, they canât have a normal distribution.

**Note:** Parametric tests canât test hypotheses about the distribution of a categorical variable, but they can involve a categorical variable as an independent variable (e.g., ANOVAs). 


## Chi-Square Goodness of Fit Test

A chi-square (Î§2) goodness of fit test is a type of Pearsonâs chi-square test. You can use it to test whether the observed distribution of a categorical variable differs from your expectations.  

The chi-square goodness of fit test is a hypothesis test. It allows you to draw conclusions about the distribution of a population based on a sample. Using the chi-square goodness of fit test, you can test whether the goodness of fit is âgood enoughâ to conclude that the population follows the distribution.

With the chi-square goodness of fit test, you can ask questions such as: Was this sample drawn from a population that has : 

Equal proportions of male and female turtles?  
Equal proportions of red, blue, yellow, green, and purple jelly beans?  
90% right-handed and 10% left-handed people?  
Offspring with an equal probability of inheriting all possible genotypic combinations (i.e., unlinked genes)?  
A Poisson distribution of floods per year?  
A normal distribution of bread prices?

### Example

Here's an example of a Chi-Square goodness of fit test to illustrate the concept more clearly:

**Scenario:** Imagine you're a marketing manager at a clothing store selling T-shirts in three sizes: small, medium, and large. You suspect that the sizes sold might not be distributed evenly based on popularity. You collect data on 100 T-shirt sales over a week and record the sizes chosen:

| Size | Frequency |
|---|---|
| Small | 28 |
| Medium | 35 |
| Large | 37 |

**Question:** Do the observed frequencies of T-shirt sizes sold deviate significantly from what we would expect if all sizes were equally popular (null hypothesis)?

**Steps:**

1. **State the null and alternative hypotheses:**
    * **Null hypothesis (H0):** The sizes sold are equally popular (33.33% each).
    * **Alternative hypothesis (Ha):** The sizes sold are not equally popular.

2. **Calculate the expected frequencies:**
    * If all sizes were equally popular, you'd expect 100 sales / 3 sizes = 33.33 sales for each size.

3. **Calculate the Chi-Square statistic:**
    * Use the formula for Chi-Square goodness of fit:
      Chi-Square = Î£ [(Observed frequency - Expected frequency)^2 / Expected frequency]
    * For each size:
        * Small: (28 - 33.33)^2 / 33.33 â 0.85
        * Medium: (35 - 33.33)^2 / 33.33 â 0.083
        * Large: (37 - 33.33)^2 / 33.33 â 0.404
    * Total Chi-Square = 0.85 + 0.083 + 0.404 â 1.33

4. **Determine the degrees of freedom and critical value:**
    * Degrees of freedom (df) = number of categories - 1 = 3 - 1 = 2
    * Choose a significance level (alpha) like 0.05. Look up the critical value for df = 2 and alpha = 0.05 in a Chi-Square distribution table. It's approximately 5.99.

5. **Compare the Chi-Square statistic to the critical value:**
    * Our calculated Chi-Square (1.33) is less than the critical value (5.99).

6. **Make a decision:**
    * Since the calculated Chi-Square is less than the critical value, we **fail to reject the null hypothesis**. There's not enough evidence to conclude that the observed frequencies deviate significantly from equal popularity.

**Interpretation:**

Based on this analysis, we don't have enough evidence to suggest that the sizes sold are not equally popular. However, this doesn't necessarily mean they are truly equal. A larger sample size or further analysis might reveal different patterns.


````python
import numpy as np

observed = np.array([28,35,37])
print("Observed values are : ",observed)

# for expected we assume all three size of shirt are equal distributed
expected = observed.sum()/len(observed)
print("Expected values are : ",expected," ",expected," ",expected)  #it can be different for each observation if assumption of null hypothesis is not equal distributed

#Formula : sum of (Ob - ex)^2 / ex

chi2_stat = 0
for i in observed:
    chi2_stat += (i-expected)**2 / expected
print("Chi-Square statistic is : ",chi2_stat)

alpha = 0.05
df = len(observed)-1
````

now decide whether reject the null hypothesis or fail to reject the null hypothesis.   
Two ways to reject or fail to reject our null hypothesis 
1. Find p value of chi-square statistic if greater than alpha, fail to reject if less than alpha, reject
2. Find critical value by finding chi square statistic of alpha with the given degree of freedom on chi square table. If chi square statistic is greater than critical value reject null hypothesis, if chi square statistic is less than critical value fail to reject the null hypothesis

````python
from scipy.stats import chi2

p_value = chi2.sf(chi2_stat, df)
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
````

**Scenario:**

Imagine you're a clothing store owner and want to see if there's a relationship between the size of T-shirts sold and the gender of the customer buying them. You collect data on 100 T-shirt sales, recording the size (Small, Medium, Large) and gender (Male, Female) of each customer.

**Question:**

Is there a statistically significant association between T-shirt size and customer gender?   
(Null hypothesis: no association, alternative hypothesis: there is an association)

**Data (contingency table):**

| Size | Male | Female | Total |
|---|---|---|---|
| Small | 20 | 15 | 35 |
| Medium | 30 | 25 | 55 |
| Large | 25 | 10 | 35 |
| Total | 75 | 50 | 125 |

**Steps:**

1. **Create a contingency table:** This table shows the frequency of each combination of size and gender.

2. **Calculate expected counts:** Assuming no association, calculate the expected counts for each cell based on marginal totals.

3. **Calculate Chi-Square statistic:** Use the formula: Î£ [(Observed frequency - Expected frequency)^2 / Expected frequency]

4. **Determine degrees of freedom and critical value:** df = (number of rows - 1) * (number of columns - 1). Choose a significance level (alpha) and look up the critical value from a Chi-Square distribution table.

5. **Compare Chi-Square statistic to critical value:** Reject the null hypothesis if Chi-Square statistic > critical value.

Unfortunately, calculating the Chi-Square statistic for independence **requires more complex calculations than can be easily performed manually**. While it's possible to understand the concepts and formula without using code, manually executing all the calculations for a real-world dataset can be tedious and error-prone.

However, I can guide you through the steps and calculations conceptually, highlighting why using code is generally recommended:

**Steps:**

1. **Contingency Table:**

This one you already have:

| Size | Male | Female | Total |
|---|---|---|---|
| Small | 20 | 15 | 35 |
| Medium | 30 | 25 | 55 |
| Large | 25 | 10 | 35 |
| Total | 75 | 50 | 125 |

2. **Expected Counts:**

For each cell, calculate the expected count assuming no association between size and gender. This involves multiplying the marginal totals (row and column sums) and dividing by the overall total:

expected value = row r * col c / N

$ \text{Expected for small-male} = \frac{{35 \times 75}}{{125}} = \frac{{2625}}{{125}} = 21 $

$ \text{Expected for small-female} = \frac{{35 \times 50}}{{125}} = \frac{{1750}}{{125}} = 14 $

$ \text{Expected for medium-male} = \frac{{55 \times 75}}{{125}} = \frac{{4125}}{{125}} = 33 $

$ \text{Expected for medium-female} = \frac{{55 \times 50}}{{125}} = \frac{{2750}}{{125}} = 22 $

$ \text{Expected for large-male} = \frac{{35 \times 75}}{{125}} = \frac{{2625}}{{125}} = 21 $

$ \text{Expected for large-female} = \frac{{35 \times 50}}{{125}} = \frac{{1750}}{{125}} = 14 $
 


3. **Individual Contributions:**

Formula : (Observed - Expected)^2 / Expected


| Size   | Gender | Observed | Expected | O-E | (O-E)^2 | (O-E)^2 / E |
|--------|--------|-------|----------|-----|---------|-------------|
| Small  | Male   | 20    | 21       | -1  | 1       | 0.0476      |
| Small  | Female | 15    | 14       | 1   | 1       | 0.0714      |
| Medium | Male   | 30    | 33       | -3  | 9       | 0.2727      |
| Medium | Female | 25    | 22       | 3   | 9       | 0.4091      |
| Large  | Male   | 25    | 21       | 4   | 16      | 0.7619      |

|--------|--------|-------|----------|-----|---------|-------------|
| Chi-Statistic  |        |       |          |     |         | 2.4229      |


4. **Chi-Square Statistic:**

Sum the individual contributions from all cells. This is your Chi-Square statistic.  
Chi-Statistic  = 2.4229

5. **Degrees of Freedom:**

df = (number of rows - 1) * (number of columns - 1) = (3-1) * (2-1) = 2

6. **Critical Value:**

Look up the critical value from a Chi-Square distribution table based on your chosen significance level (0.05) and degrees of freedom (2).  
critical value = 5.991

7. **Interpretation:**
Compare your Chi-Square statistic to the critical value. If your statistic is **greater than** the critical value, you **reject the null hypothesis** (no association) and conclude a significant association between size and gender. Otherwise, you **fail to reject the null hypothesis**.

chi-statistic < critical value : So we fail to reject the null hypothesis



````python
import pandas as pd
from scipy.stats import chi2_contingency

# Data
sizes = ["Small", "Medium", "Large"]
genders = ["Male", "Female"]
data = [[20, 15], [30, 25], [25, 10]]

# Create DataFrame
df = pd.DataFrame(data, index=sizes, columns=genders)

# Calculate Chi-Square statistic
chi2_stat, p_value, dof, expected = chi2_contingency(df)

# Print results
print("Contingency table:")
print(df)
print("Expected counts:")
print(expected)
print("Chi-Square statistic:", chi2_stat)
print("p-value:", p_value)
print("Degrees of freedom:", dof)

# Interpretation (example for alpha = 0.05)
if p_value <= 0.05:
    print("Reject null hypothesis. There is a significant association between size and gender.")
else:
    print("Fail to reject null hypothesis. Not enough evidence to conclude an association.")
````

# Analysis of Variance (ANOVA)

ANOVA, which stands for Analysis of Variance, is a statistical test used to analyze the difference between the means of more than two groups.

**Two Types**  
A one-way ANOVA uses one independent variable (factor), while a two-way ANOVA uses two independent variables (factors).  
***note***: The independent variable (factor) should have at least three levels (i.e. at least three different groups or categories)

## Assumptions of ANOVA

The assumptions of the ANOVA test are the same as the general assumptions for any parametric test:

***Independence of observations***:   
the data were collected using statistically valid sampling methods, and there are no hidden relationships among observations. If your data fail to meet this assumption because you have a confounding variable that you need to control for statistically, use an ANOVA with blocking variables.  
***Normally-distributed response variable***:  
The values of the dependent variable follow a normal distribution.   
***Absence of outliers***:  
outliers must be removed from the dataset.  
***Homogeneity of variance***:  
The variation within each group being compared is similar for every group. If the variances are different among the groups, then ANOVA probably isnât the right fit for the data.


## When to use a one-way ANOVA

Use a one-way ANOVA when you have collected data about one categorical independent variable and one quantitative dependent variable. The independent variable should have at least three levels (i.e. at least three different groups or categories).

ANOVA tells you if the dependent variable changes according to the level of the independent variable. 

### Example:

Your independent variable is social media use, and you assign groups to ***low***, ***medium***, and ***high*** levels of social media use to find out if there is a difference in hours of ***sleep per night***.  
Your independent variable is brand of soda, and you collect data on ***Coke***, ***Pepsi***, ***Sprite***, and ***Fanta*** to find out if there is a difference in the ***price per 100ml***.  
You independent variable is type of fertilizer, and you treat crop fields with mixtures ***1***, ***2*** and ***3*** to find out if there is a difference in ***crop yield***.  
The null hypothesis (H0) of ANOVA is that there is no difference among group means. The alternative hypothesis (Ha) is that at least one group differs significantly from the overall mean of the dependent variable.

If you only want to compare two groups, use a t test instead.


## When to use Two-Way ANOVA

A two-way ANOVA is used to estimate how the mean of a quantitative variable changes according to the levels of two categorical variables. Use a two-way ANOVA when you want to know how two independent variables (factors), in combination, affect a dependent variable.

### Example:
You are researching which type of fertilizer and planting density produces the greatest crop yield in a field experiment. You assign different plots in a field to a combination of fertilizer type (1, 2, or 3) and planting density (1=low density, 2=high density), and measure the final crop yield in bushels per acre at harvest time.
You can use a two-way ANOVA to find out if fertilizer type and planting density have an effect on average crop yield


````python
import seaborn as sns
import pandas as pd
from scipy import stats

# Load the Titanic dataset
df = sns.load_dataset("titanic")
df.head()
````

````python
# Group data by Pclass and calculate means for survival
pclass_survival_means = df.groupby("pclass")["survived"].mean()
pclass_survival_means
````

````python
# Perform One-way ANOVA test
f_value, p_value = stats.f_oneway(
    df[df["pclass"] == 1]["survived"],
    df[df["pclass"] == 2]["survived"],
    df[df["pclass"] == 3]["survived"],
)

# Print the results
print("F-statistic:", f_value)
print("p-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("There is a statistically significant difference in survival rates between Pclass groups.")
else:
    print("There is not enough evidence to conclude a statistically significant difference in survival rates between Pclass groups.")
````

````python
#Two-way Anova
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Load the Titanic dataset
titanic_df = sns.load_dataset("titanic")

# Filter relevant columns
filtered_df = titanic_df[['sex', 'pclass', 'fare']]  # Correct column names

# Fit the two-way ANOVA model (using appropriate column names)
model = ols('fare ~ sex + pclass', data=filtered_df).fit()

# Perform the hypothesis test
anova_table = sm.stats.anova_lm(model, typ=2)

# Display the ANOVA table
print(anova_table)
````

**Interpretation**:

sum_sq: This column represents the sum of squares, which measures the variance within each group. A higher sum of squares indicates greater variability within the group.  
df: This column represents the degrees of freedom, which is a measure of the number of independent pieces of information in each group.  
F: This column represents the F-statistic, which is a test statistic used to compare the variance between groups to the variance within groups. A higher F-statistic indicates a stronger effect of the group on the dependent variable.  
PR(>F): This column represents the p-value, which is the probability of observing an F-statistic as extreme as the one calculated, assuming there is no effect of the group on the dependent variable. A lower p-value indicates stronger evidence against the null hypothesis (i.e., there is a statistically significant effect of the group on the dependent variable).  

**Looking at the table**:

Sex: The F-statistic (15.903249) and p-value (7.214818e-05) suggest that there is a statistically significant effect of sex on the dependent variable. This means that the variances between the two sexes are significantly different.
Pclass: The F-statistic (363.848060) and p-value (3.000415e-68) provide even stronger evidence for a statistically significant effect of pclass on the dependent variable. This indicates that the variances between the different pclass groups are highly statistically significant.
